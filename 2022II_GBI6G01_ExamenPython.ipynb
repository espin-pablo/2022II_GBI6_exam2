````python
from Bio import SeqIO

def fasta_to_txt(fasta_file, txt_file):
    sequences = SeqIO.parse(fasta_file, "fasta")
    with open(txt_file, "w") as f:
        for seq in sequences:
            f.write(str(seq.seq) + "\n")

fasta_to_txt("data/coati.fasta", "cadenadetexto.txt")
````

````python
from Bio import SeqIO

def leer_secuencias_fasta(nombre_archivo):
    secuencias = []
    for secuencia in SeqIO.parse(nombre_archivo, 'fasta'):
        secuencias.append(str(secuencia.seq))
    return secuencias

qwe = leer_secuencias_fasta('data/coati.fasta')
qwe
````

````python
from Bio.Seq import Seq

dna_sequences = qwe
protein_sequences = [ ]

for dna_seq in dna_sequences:
    seq = Seq(dna_seq)
    protein_seq = seq.translate()
    protein_sequences.append(protein_seq)

print(protein_sequences)

protein_sequences = [item.replace("*", "\n") for item in protein_sequences]
with open("protein_sequences.txt", "w") as prote:
    for item in protein_sequences:
        prote.write(str(item) + "\n")
````

````python
from Bio.SeqUtils import molecular_weight

amino_acid_sequence = "CPNRISRSREFHF"

# Cálculo del peso molecular
molecular_weight = molecular_weight(amino_acid_sequence, "protein")

# Impresión del resultado
print("El peso molecular de la proteína es:", molecular_weight, "Da")
````
